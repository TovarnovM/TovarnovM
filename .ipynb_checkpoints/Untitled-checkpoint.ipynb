```python
y = b.get_y0()
ts, dt = np.linspace(0,3,1000000,retstep=True)
ys = []
for t in tqdm(ts):
    ys.append(y)
    k1 = b.get_dydt(t, y)
    k2 = b.get_dydt(t + dt/2, y + dt/2 * k1)
    k3 = b.get_dydt(t + dt/2, y + dt/2 * k2)
    k4 = b.get_dydt(t + dt,   y + dt * k3)
    y1 = y + dt/6 * (k1 + 2* k2 + 2* k3 + k4)
    y = y + dt * b.get_dydt(t + dt/2, y1)
```

```python
@interact(ind=ipywidgets.IntSlider(min=0,max=len(ys)-1,step=10, value=0))
def plotty(ind):
    t = ts[ind]
    y = ys[ind]
#     print(y)
    b.synch(t, y)
    fig, ax = plt.subplots()
    b.plot(ax)
    b2.plot(ax)
    plt.grid(True)
    ax.set_aspect('equal', adjustable='box')
    ax.set_xlim(-10, 20)
    ax.set_ylim(-15,3)
    plt.title(f'{t}')
    plt.show()
```
